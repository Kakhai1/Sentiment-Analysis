```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Resources/Credit Score Classification Dataset.csv")
```

```python
hist, bins, _ = plt.hist(df['Age'], bins=4, color='blue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Annotate the bars with their respective counts
for i in range(len(bins) - 1):
    plt.annotate(str(int(hist[i])), (bins[i] + 0.5, hist[i]), ha='center', va='bottom')

# Display the histogram
plt.savefig('EDA/Age Distribution.png', dpi=300)

plt.show()
print('Minimum Age: ' + str(df['Age'].min()))
print('Maximum Age: ' + str(df['Age'].max()))
```

```python
gender_counts = df['Gender'].value_counts()

# Create a bar plot for the 'Marital Status' column
plt.bar(gender_counts.index, gender_counts)
plt.title('gender_counts Distribution')
plt.xlabel('gender_counts')
plt.ylabel('Count')
for i, count in enumerate(gender_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')
# Rotate the x-axis labels for better readability if needed
plt.xticks(rotation=45)

plt.savefig('EDA/gender_counts.png', dpi=300)

# Display the bar plot
plt.show()
```

```python
hist, bins, _ = plt.hist(df['Income'], bins=7, color='blue', edgecolor='black')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')

# Annotate the bars with their respective counts
for i in range(len(bins) - 1):
    plt.annotate(str(int(hist[i])), (bins[i] + 0.5, hist[i]), ha='center', va='bottom')
plt.savefig('EDA/Income Distribution.png', dpi=300)

# Display the histogram
plt.show()
print('Minimum Income: ' + str(df['Income'].min()))
print('Maximum Income: ' + str(df['Income'].max()))
```
